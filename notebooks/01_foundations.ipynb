## Series와 DataFrame

- Series = 1차원 데이터 + 인덱스(라벨). 엑셀의 한 열
- DataFrame = 2차원 표. 여러 Series(열)가 같은 인덱스를 공유

```python
import pandas as pd
import numpy as np

print(pd.__version__) # 버전 확인
```

```python
# 리스트 만들기
s = pd.Series([1,2,3,4])
print(s)
```

```python
s2 = pd.Series([90, 85, 88], index=["국어", "영어", "수학"])
print(s2)

print(s2["영어"])
print(s2.values)
print(s2.index)
```

```python
# DataFrame 만들어보기

df = pd.DataFrame({
    "이름": ["철수", "영희", "민수"],
    "나이": [25, 30, 28],
    "도시": ["서울", "부산", "서울"],
})
print(df)
```
